## Visualize the data

### Subtask:
Create various plots (e.g., bar plots, box plots, violin plots) using Seaborn and Matplotlib to explore relationships between features and survival.


**Reasoning**:
Create various plots using Seaborn and Matplotlib to explore relationships between features and survival.



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y='survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='survived', y='age', data=df)
plt.title('Age Distribution for Survivors vs. Non-Survivors')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()

plt.figure(figsize=(8, 6))
sns.violinplot(x='survived', y='fare', data=df)
plt.title('Fare Distribution for Survivors vs. Non-Survivors')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='sibsp', y='survived', data=df)
plt.title('Survival Rate by Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='parch', y='survived', data=df)
plt.title('Survival Rate by Number of Parents/Children Aboard')
plt.xlabel('Number of Parents/Children Aboard')
plt.ylabel('Survival Rate')
plt.show()
```
